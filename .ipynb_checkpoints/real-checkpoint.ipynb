```python
agent.dynamic_treeify()
agent.reset_to_root()
curt_node = agent.ROOT
paths = []
leafsnode=[]
def construct_paths(root, path):
    if root:
        path.append(root)
            #print(path)
        if root.is_leaf():  # 当前节点是叶子节点
            leafsnode.append(root)
        else:
            construct_paths(root.kids[0], path)
            construct_paths(root.kids[1], path)
construct_paths(curt_node,[])        
paths=[]
for i in leafsnode:
    path=[]
    cc=i
    while cc is not None:

        if cc.is_good_kid():
            path.append((cc.parent,0))
        else:
            path.append((cc.parent,1))
        cc=cc.parent
    path=path[:-1]   
    path=path[::-1]
    paths.append(path)
bili=[i.get_uct() for i in leafsnode]
```

```python
low_b=[1]*72
up_b=[0]*72
for i in range(1):
    if leafsnode[i].is_good_kid():
        ll=leafsnode[i].propose_samples_bo1(100000,1, paths[i],agent.lb, agent.ub, agent.samples)
        for j in range(72):
            #print(ll[:,j].min())
            if ll[:,j].min()<low_b[j]:
                low_b[j]=ll[:,j].min()
            #print(ll[:,j].max())
            if ll[:,j].max()>up_b[j]:
                up_b[j]=ll[:,j].max()

low_b=np.array(low_b)
up_b=np.array(up_b)
def r2(x):
    return x*(up_b-low_b)+low_b

def b2(x):
    return (x-low_b)/(up_b-low_b)
```

```python
import torch
import numpy as np
import torch.nn as nn
torch.set_default_tensor_type(torch.FloatTensor)
class SCF_model(torch.nn.Module):
    def __init__(self, input_width, layer_width):
        super(SCF_model, self).__init__()
        self.linear_in = torch.nn.Linear(input_width, layer_width)
        self.bc1=torch.nn.BatchNorm1d(layer_width, momentum=0.9)
        self.lru1=  nn.LeakyReLU(0.2)
        
        self.linear2 = torch.nn.Linear(layer_width, 192)
        self.bc2=torch.nn.BatchNorm1d(192, momentum=0.9)
        self.lru2=  nn.LeakyReLU(0.2)
        
        
        kernel_size = (4,3)
        self.m1 = nn.ConvTranspose2d(4, 256, (2, 1), stride=(2, 1))
        self.bc3=torch.nn.BatchNorm2d(256, momentum=0.9)
        self.lru3=  nn.LeakyReLU(0.2)
        #self.m1=nn.ConvTranspose2d(4, 33, 3, kernel_size=2, stride=(2,1))
        
        self.m2 = nn.ConvTranspose2d(256, 64, (2, 1), stride=(2, 1))
        self.bc4=torch.nn.BatchNorm2d(64, momentum=0.9)
        self.lru4=  nn.LeakyReLU(0.2)
        
        self.m3=nn.Conv2d(64,32, kernel_size, stride=1, padding=0, dilation=1, groups=1,bias=True)
        
        
        
        self.linear1 = torch.nn.Linear(4032, 1024)
        self.linear5 = torch.nn.Linear(1024, 1)
       
    def forward(self, x):
        #kernel_size = (4,3)
        y = self.linear_in(x) # fully connect layer
        y=self.bc1(y)
        y=self.lru1(y)
        
        y =self.linear2(y)
        y=self.bc2(y)
        y=self.lru2(y)
        
        y=y.reshape(-1,4,3,16)
        
        
        y=self.m1(y)
        y=self.bc3(y)
        y=self.lru3(y)
        
        y=self.m2(y)
        y=self.bc4(y)
        y=self.lru4(y)
        
        y=self.m3(y)
        y=torch.tanh(y)
        shape1=y.shape[0]
        y=y.reshape(shape1,-1)
        
        y=torch.sigmoid(self.linear1(y))
        y=torch.sigmoid(self.linear5(y))*0.01+1/2000
        return y
net=SCF_model(72,1024).cuda()
```

# PPO编写

```python
import torch.nn as nn
from tqdm import tqdm
import numpy as np
import torch
import collections
import random

class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = collections.deque(maxlen=capacity) 

    def add(self, state, action, reward, next_state, done): 
        self.buffer.append((state, action, reward, next_state, done)) 

    def sample(self, batch_size): 
        transitions = random.sample(self.buffer, batch_size)
        state, action, reward, next_state, done = zip(*transitions)
        return np.array(state), action, reward, np.array(next_state), done 

    def size(self): 
        return len(self.buffer)

def moving_average(a, window_size):
    cumulative_sum = np.cumsum(np.insert(a, 0, 0)) 
    middle = (cumulative_sum[window_size:] - cumulative_sum[:-window_size]) / window_size
    r = np.arange(1, window_size-1, 2)
    begin = np.cumsum(a[:window_size-1])[::2] / r
    end = (np.cumsum(a[:-window_size:-1])[::2] / r)[::-1]
    return np.concatenate((begin, middle, end))

def train_on_policy_agent(env, agent, num_episodes):
    return_list = []
    for i in range(10):
        with tqdm(total=int(num_episodes/10), desc='Iteration %d' % i) as pbar:
            for i_episode in range(int(num_episodes/10)):
                episode_return = 0
                transition_dict = {'states': [], 'actions': [], 'next_states': [], 'rewards': [], 'dones': []}
                state = env.reset()
                done = False
                while not done:
                    action = agent.take_action(state)
                    next_state, reward, done, _ = env.step(action)
                    transition_dict['states'].append(state)
                    transition_dict['actions'].append(action)
                    transition_dict['next_states'].append(next_state)
                    transition_dict['rewards'].append(reward)
                    transition_dict['dones'].append(done)
                    state = next_state
                    episode_return += reward
                return_list.append(episode_return)
                agent.update(transition_dict)
                if (i_episode+1) % 10 == 0:
                    pbar.set_postfix({'episode': '%d' % (num_episodes/10 * i + i_episode+1), 'return': '%.3f' % np.mean(return_list[-10:])})
                pbar.update(1)
    return return_list

def train_off_policy_agent(env, agent, num_episodes, replay_buffer, minimal_size, batch_size):
    return_list = []
    for i in range(10):
        with tqdm(total=int(num_episodes/10), desc='Iteration %d' % i) as pbar:
            for i_episode in range(int(num_episodes/10)):
                episode_return = 0
                state = env.reset()
                done = False
                while not done:
                    action = agent.take_action(state)
                    next_state, reward, done, _ = env.step(action)
                    replay_buffer.add(state, action, reward, next_state, done)
                    state = next_state
                    episode_return += reward
                    if replay_buffer.size() > minimal_size:
                        b_s, b_a, b_r, b_ns, b_d = replay_buffer.sample(batch_size)
                        transition_dict = {'states': b_s, 'actions': b_a, 'next_states': b_ns, 'rewards': b_r, 'dones': b_d}
                        agent.update(transition_dict)
                return_list.append(episode_return)
                if (i_episode+1) % 10 == 0:
                    pbar.set_postfix({'episode': '%d' % (num_episodes/10 * i + i_episode+1), 'return': '%.3f' % np.mean(return_list[-10:])})
                pbar.update(1)
    return return_list


def compute_advantage(gamma, lmbda, td_delta):
    td_delta = td_delta.detach().numpy()
    advantage_list = []
    advantage = 0.0
    for delta in td_delta[::-1]:
        advantage = gamma * lmbda * advantage + delta
        advantage_list.append(advantage)
    advantage_list.reverse()
    return torch.tensor(advantage_list, dtype=torch.float)
def activatet(x):
    return torch.tanh(x)

def orthogonal_init(layer, gain=1.0):
    nn.init.orthogonal_(layer.weight, gain=gain)
    nn.init.constant_(layer.bias, 0)

class PolicyNetContinuous(torch.nn.Module):
    def __init__(self, state_dim, hidden_dim, action_dim):
        super(PolicyNetContinuous, self).__init__()
        self.fc1 = torch.nn.Linear(state_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = torch.nn.Linear(hidden_dim, hidden_dim)
        
        self.fc4 = torch.nn.Linear(hidden_dim, hidden_dim)
        self.fc5 = torch.nn.Linear(hidden_dim, hidden_dim)

        self.fc_mu = torch.nn.Linear(hidden_dim, action_dim)
        self.fc_std = torch.nn.Linear(hidden_dim, action_dim)
        #self.m=activatet
        self.m=nn.SiLU()
        self.rand1=0.2
        #self.a=torch.tensor([0.049853291779000004,0.1,0.1])
        #self.b=torch.tensor([0.05020000000000001,0,0])
        orthogonal_init(self.fc1)
        orthogonal_init(self.fc2)
        orthogonal_init(self.fc3)
        orthogonal_init(self.fc4)
        orthogonal_init(self.fc5)
        orthogonal_init(self.fc_mu, gain=0.01)
        orthogonal_init(self.fc_std, gain=0.01)
        
        

    def forward(self, x):
        x = self.m(self.fc1(x))
        x1 = self.m(self.fc2(x))
        x1 = self.m(self.fc3(x1))
        x2 = self.m(self.fc2(x))
        x2 = self.m(self.fc3(x2))
        mu =  0.6*torch.tanh(self.fc_mu(x1))
        std = self.rand1*F.softplus(self.fc_std(x2))
        return mu, std
    def setrand1(self,k2):
        self.rand1=k2
    
    
    
class ValueNet(torch.nn.Module):
    def __init__(self, state_dim, hidden_dim):
        super(ValueNet, self).__init__()
        self.fc1 = torch.nn.Linear(state_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = torch.nn.Linear(hidden_dim, hidden_dim)
        self.m=nn.SiLU()
        #self.m=activatet
        self.fc_out = torch.nn.Linear(hidden_dim, 1)
        
        orthogonal_init(self.fc1)
        orthogonal_init(self.fc2)
        orthogonal_init(self.fc3)
        orthogonal_init(self.fc_out, gain=0.01)
        


    def forward(self, x):
        x = self.m(self.fc1(x))
        x = self.m(self.fc2(x))
        x = self.m(self.fc3(x))
        return self.fc_out(x)
    
    



class PPOContinuous:
    ''' 处理连续动作的PPO算法 '''
    def __init__(self, state_dim, hidden_dim, action_dim, actor_lr, critic_lr,
                 lmbda, epochs, eps, gamma, device):
        self.actor = PolicyNetContinuous(state_dim, hidden_dim,
                                         action_dim).to(device)
        self.critic = ValueNet(state_dim, hidden_dim).to(device)
        self.actor_optimizer = torch.optim.Adam(self.actor.parameters(),
                                                lr=actor_lr) #tr2
        self.critic_optimizer = torch.optim.Adam(self.critic.parameters(),
                                                 lr=critic_lr)#tr2
        self.gamma = gamma
        self.lmbda = lmbda
        self.epochs = epochs
        self.eps = eps
        self.device = device
        self.lr_a=actor_lr
        self.lr_c=critic_lr

    def take_action(self, state,l1=True):
        state = torch.tensor(state, dtype=torch.float).to(self.device)
        mu, sigma = self.actor(state)
        if l1:
            action_dist = torch.distributions.Normal(mu, sigma)
            action = action_dist.sample()
        else:
            action=mu.detach()
            
        return action.numpy()

    def update(self, transition_dict,total_steps):
        states = torch.tensor(transition_dict['states'],
                              dtype=torch.float).to(self.device)
        actions = torch.tensor(transition_dict['actions'],
                               dtype=torch.float).view(-1, 72).to(self.device)
        rewards = torch.tensor(transition_dict['rewards'],
                               dtype=torch.float).view(-1, 1).to(self.device)
        next_states = torch.tensor(transition_dict['next_states'],
                                   dtype=torch.float).to(self.device)
        dones = torch.tensor(transition_dict['dones'],
                             dtype=torch.float).view(-1, 1).to(self.device)
        #rewards = rewards  # 和TRPO一样,对奖励进行修改,方便训练
        td_target = rewards + self.gamma * self.critic(next_states) * (1 -
                                                                       dones)
        td_delta = td_target - self.critic(states)
        advantage = compute_advantage(self.gamma, self.lmbda,
                                               td_delta.cpu()).to(self.device)
        mu, std = self.actor(states)
        action_dists = torch.distributions.Normal(mu.detach(), std.detach())
        # 动作是正态分
        old_log_probs = action_dists.log_prob(actions)
        self.lr_decay(total_steps)#tirck
        for _ in range(self.epochs):
            mu, std = self.actor(states)
            action_dists = torch.distributions.Normal(mu, std)
            log_probs = action_dists.log_prob(actions)
            ratio = torch.exp(log_probs - old_log_probs)
            surr1 = ratio * advantage
            surr2 = torch.clamp(ratio, 1 - self.eps, 1 + self.eps) * advantage
            actor_loss = torch.mean(-torch.min(surr1, surr2))
            critic_loss = torch.mean(
                F.mse_loss(self.critic(states), td_target.detach()))
            self.actor_optimizer.zero_grad()
            self.critic_optimizer.zero_grad()
            actor_loss.backward()
            critic_loss.backward()
            torch.nn.utils.clip_grad_norm_(self.actor.parameters(), 0.3) #tr1
            torch.nn.utils.clip_grad_norm_(self.critic.parameters(), 0.3) #tr1
            self.actor_optimizer.step()
            self.critic_optimizer.step()
            
    def lr_decay(self, total_steps):#tirck
        lr_a_now = self.lr_a * (1 - total_steps / 10000)
        lr_c_now = self.lr_c * (1 - total_steps / 10000)
        for p in self.actor_optimizer.param_groups:
            p['lr'] = lr_a_now
        for p in self.critic_optimizer.param_groups:
            p['lr'] = lr_c_now
            
    
    def adjust_learning_rate(self,optimizer, total_steps,decay_rate=.9):
        for param_group in optimizer.param_groups:
            param_group['lr'] = param_group['lr'] * decay_rate
```

```python
def get_reward(state1,state):
    mark=0
    for i in r2(state):
        if i<0 or i>1:
            mark=1
            break
    if mark != 1:
        return eqn(state)-eqn(state1)
    else:
        temp=0
        for i in r2(state):
            if i<0:
                temp=temp+abs(0-i)
            if i>1:
                temp=temp+abs(i-1)
            
        return -10*temp
```

```python
def step(action,state,kk):
    next_state=state+action
    reward=get_reward(state.flatten(),next_state.flatten())
    done=0
    if kk>20:
        done=1
    return next_state,reward,done
```
